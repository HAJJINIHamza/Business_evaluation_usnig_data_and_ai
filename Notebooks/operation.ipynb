# Enlever la normalization des colonnes EBITDA et EBITDA marg (%) du data echantillon test

```python
import pandas as pd
import pickle 
```

### Inverser la normalization

```python
#Importation du normalizeur
with open ( 'Models_V2/normalizeur.pkl', 'rb' ) as f:
    normalizeur = pickle.load (f)
```

```python
#Importation de la data
data = pd.read_excel ( 'Data Finpack V2/resultats_test_echantillon/resultats_test_echantillon.xlsx', sheet_name = 'Six Flags Entertainment Corp.', index_col= 0 )
data.head()
```

```python
#Selectionne des colonnes  à inverser la normalisation
data_ = data [['Sector', 'Country', 'EBITDA - Last', 'EBITDA Marg. (%) - Last']]
data_.head ()
```

```python
#Application du normalizeur
data_2 = normalizeur.inverse_transform ( data_ )
data_2 = pd.DataFrame ( data_2, columns = data_.columns, index = data_.index )
data_2
```
