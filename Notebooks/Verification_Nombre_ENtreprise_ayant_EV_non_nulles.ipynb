# Verification du nombre d'entreprise ayant des valeurs EV non nulles 2

Ce notebook sert a verfier le nombre d'entreprise ayant des valeurs non nulles dans les colonnes EV/EBITDA, EV/Sales, ...

### **Packages** 

```python
import pandas as pd
import numpy as np
```

### **Data d'input** 

```python
data_sans_na_sans_out = pd.read_excel ( 'Data Finpack V2/data_finpack_V2_sans_na_sans_out.xlsx', index_col = 0 )
data_sans_na_sans_out
```

```python
big_data = pd.read_excel ( 'Data Finpack V2/data_finpack_V2.xlsx' )
big_data
```

```python
ev_ebitda = big_data [ 'EV/EBITDA - Last' ]
ev_ebit = big_data ['EV/EBIT - Last']
ev_sales = big_data ['EV/Sales - Last']
price_earning = big_data ['P/E - Last']
price_book = big_data ['P/Book Value - Last']
```

```python
data_sans_na_sans_out_ev = pd.merge ( data_sans_na_sans_out, ev_ebitda, right_index = True, left_index = True )
data_sans_na_sans_out_ev = pd.merge ( data_sans_na_sans_out_ev, ev_ebit, right_index = True, left_index = True )
data_sans_na_sans_out_ev = pd.merge ( data_sans_na_sans_out_ev, ev_sales, right_index = True, left_index = True )
data_sans_na_sans_out_ev = pd.merge ( data_sans_na_sans_out_ev, price_earning, right_index = True, left_index = True )
data_sans_na_sans_out_ev = pd.merge ( data_sans_na_sans_out_ev, price_book, right_index = True, left_index = True )
data_sans_na_sans_out_ev
```

```python
colonne1 = 'EV/EBITDA - Last'
colonne2 = 'EV/EBIT - Last'
colonne3 = 'EV/Sales - Last'
colonne4 = 'P/E - Last'
colonne5 = 'P/Book Value - Last'
print ('Nombre de valeurs non nulles dans la colonne ',  colonne1,  ', est =', data_sans_na_sans_out_ev_ebitda.shape[0] - data_sans_na_sans_out_ev [ colonne1 ].isna().sum())
print ('Nombre de valeurs non nulles dans la colonne ',  colonne2,  ', est =', data_sans_na_sans_out_ev_ebitda.shape[0] - data_sans_na_sans_out_ev [ colonne2 ].isna().sum())
print ('Nombre de valeurs non nulles dans la colonne ',  colonne3,  ', est =', data_sans_na_sans_out_ev_ebitda.shape[0] - data_sans_na_sans_out_ev [ colonne3 ].isna().sum())
print ('Nombre de valeurs non nulles dans la colonne ',  colonne4,  ', est =', data_sans_na_sans_out_ev_ebitda.shape[0] - data_sans_na_sans_out_ev [ colonne4 ].isna().sum())
print ('Nombre de valeurs non nulles dans la colonne ',  colonne5,  ', est =', data_sans_na_sans_out_ev_ebitda.shape[0] - data_sans_na_sans_out_ev [ colonne5 ].isna().sum())
```
