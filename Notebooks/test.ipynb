#### Test 1

```python
#Data
X,y = make_blobs (n_samples = 1000, centers = 5)
```

```python
fig = px.scatter ( x = X[:,0], y = X[:,1], width = 700, height = 500 )
fig.show()
```

```python
#Diagramme d'ELBOW et diagramme de la silhouette score 
inertia = []
silhouette = []
davies_bouldin = []
Nbr = [ i for i in range (2, 20) ]
for n in Nbr:
    model = KMeans ( n_clusters = n, n_init = 13 )
    model.fit( X )
    inertia.append (model.inertia_)
    silhouette.append ( silhouette_score( X, model.labels_ ) )
    davies_bouldin.append ( davies_bouldin_score(X, model.labels_) )
    print ( 'Trmin pour n_cluster :', n )
```

```python
#Diagrmme d'elbow
fig5 = px.line( x = range (2, len( inertia )+2), y = inertia, width = 700, height =  500  )
fig5.show()
```

```python
#Diagramme silhouette
fig6 = px.line( x = range (2, len( silhouette )+2), y = silhouette, width = 700, height =  500  )
fig6.show()
```

```python
#Diagramme Davies bouldin
fig8 = px.line ( x = [i for i in  range ( 2, len (davies_bouldin)+2 )], y = davies_bouldin, width = 700, height =  500 )
fig8.show()
```

```python
#Normalisation des donnes
X_norm = RobustScaler ().fit_transform (X)
```

```python
X_1 = np.vstack ((X, X_norm))
len (X_1)
```

```python
silhouette = []
eps_range = np.arange (0.2, 3.1, 0.1)
for eps in eps_range:
    model = DBSCAN (eps = eps)
    model.fit (X)
    silhouette.append ( silhouette_score ( X, model.labels_ ) )
    print ('trmin pour eps : ', eps)
```

```python
#Diagramme de la silhouette
fig14 = px.line( x = eps_range, y=silhouette, width = 700, height = 500 )
fig14.update_layout (  title = 'Silhouette' )
fig14.show()
```

```python
eps = 0.8
model_DBSCAN = DBSCAN ( eps = eps )
model_DBSCAN.fit ( X )
model_DBSCAN_labels = model_DBSCAN.labels_
model_DBSCAN_labels = pd.DataFrame( model_DBSCAN_labels, columns = ["labels"] )
model_DBSCAN_labels.value_counts()
```

```python
#calcule de silhouette et indice davies bouldin
print ( 'Silhouette score du model DBSCAN pour eps = :', eps  , ' est :' , silhouette_score(X, model_DBSCAN.labels_) )
print ( '''L'indice  de Davies Bouldin du model DBSCAN pour eps =''', eps, 'est ', davies_bouldin_score(X, model_DBSCAN.labels_))
#print ('Inertia du model MiniBatchKMeans n_clusters = :', n_clusters , ' est :', model_MiniBatchKMeans.inertia_)
```

```python
#X1
x1 = pd.ExcelFile ( 'Holcim swiss comps.xls' )
x1
```

```python
#Boocle sur les classeur
dic = {}
for i in range (0, len( x1.sheet_names)  ):    
    #le nom du classeur
    nom_classeur = x1.sheet_names[i]
    nom_classeur
    #Importation des donnes du classeur
    data_classeur = x1.parse ( nom_classeur, header = 3 )
    data_classeur
    #Supprimer les lignes au secteur est nulles
    data_classeur_1 = data_classeur.drop ( data_classeur.index[ data_classeur['Sector'].isna() ] )
    dic[i] = data_classeur_1

```

```python
dic
```

```python
#Listes des df
l = []
for i in range (len(dic)):
    l.append (dic[i])
l
```

```python
fin = pd.concat (l)
fin
```

```python
#Initialison les index
fin_ = fin.reset_index(drop = True)
fin_
```

```python
#Suppression des doublons 
data_finale = fin_.drop ( fin_.index [fin_['Company Name'].duplicated()])
data_finale
```

```python
#Initialison les index
data_finale = data_finale.reset_index(drop = True)
data_finale
```

```python
data_finale.to_excel ( 'Holcim swiss comps V2 sans duppliqu.xlsx' )
```
